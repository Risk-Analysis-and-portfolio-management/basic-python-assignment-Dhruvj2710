```python
#QUESTION 3
def analyze_portfolio(ticker_arr,weight_arr):
    portfolio_return=0
    for ticker,weight in zip(ticker_arr,weight_arr):
        df_ticker=yf.download(ticker, start="2023-12-17",end="2024-12-17")
        df_ticker_close=df_ticker['Adj Close'] 
        daily_returns=df_ticker_close.pct_change()
        weighted_return=daily_returns*weight
        portfolio_return=weighted_return.sum(axis=1)
    return portfolio_return

ticker_arr=['AAPL','MSFT','GOOGL','TSLA', 'NVDA']
weight_arr=[0.3, 0.2, 0.2, 0.2,0.1]
y=analyze_portfolio(ticker_arr,weight_arr)
portfolio_finalreturn=100000*(y+1).cumprod()
final_return=portfolio_finalreturn.tail(1)
print("The total value of the portfolio at the end of the term:")
print(final_return)
     
```
